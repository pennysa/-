```python
#plot normal distribution
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

x = np.linspace(-5, 5, 100)
mean = 0
std_dev = 1
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, label='Normal Distribution', color='blue')
plt.fill_between(x, pdf, where=(x > -1) & (x < 1), color='lightblue', alpha=0.5, label='Area between -1 and 1')
plt.title('Normal Distribution with Shaded Area')
plt.xlabel('x')         
plt.ylabel('Probability Density')
plt.legend()
plt.grid()  
plt.show()
```

```python
#plot 5 normal distributions with different means and std deviations
import matplotlib.pyplot as plt         
import numpy as np
from scipy.stats import norm
import scipy.stats as stats

x = np.linspace(-10, 10, 1000)
means = [0, 2, -2, 4, -4]
std_devs = [1, 0.5, 1.5, 0.8, 1.2]
colors = ['#FF8C8C', '#5CB85C', '#5BC0EB', '#FFD700', '#A084CA']

plt.figure(figsize=(7,3)) # Set figure size

for mean, std_dev, color in zip(means, std_devs, colors):
    pdf = stats.norm.pdf(x, mean, std_dev)
    plt.plot(x, pdf, label=rf'$\mu$={mean}, $\sigma$={std_dev}', color=color)
    plt.fill_between(x, pdf, where=(x > mean - std_dev) & (x < mean + std_dev), color=color, alpha=0.3)

plt.title('Multiple Normal Distributions with Shaded Areas')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid() # Add grid 格線
plt.show() # Show the plot
```
